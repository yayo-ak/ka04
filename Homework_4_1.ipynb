### Problem 1

6面のサイコロを10回振るシミュレーションを行い, 5の目がでる確率を求めてください.  
ただし, ランダム数の生成には *random* ライブラリーを用いると便利です (https://note.nkmk.me/python-random-randrange-randint/).

```python
import random
sai=[random.randint(1,6) for i in range(10)]

print(sai.count(5)/10)
    
```

```python
import numpy as np
from numpy.random import randint

rolls = randint(1, 7, (10))
print(np.count_nonzero(rolls == 5) / 10)
```

### Problem 2

6面のサイコロを10,000回振るシミュレーションを行い, その結果をヒストグラムとして表示してください.  
ヒストグラムなどのグラフ描画には, *matplotlib* ライブラリーがオススメです (https://qiita.com/supersaiakujin/items/be4a78809e7278c065e6).

```python
import matplotlib.pyplot as plt
import random
import numpy as np
sai=[random.randint(1,6) for i in range(10000)]

plt.hist(sai, bins=np.arange(7)+0.5, ec='black', rwidth=0.7)
plt.title('Histogram for 10000 Die Rolls')
plt.xlabel("Outcomes")
plt.ylabel("Frequency")
plt.show()
```

### Problem 3

N個のサイコロを同時に振ったときのサイコロの目の合計を記録する作業をM回繰り返すとします(Mのデフォルト値は10,000).  
合計の散らばりをヒストグラムとして表してください.  

```python
import random
import matplotlib.pyplot as plt
import numpy as np
def rollAndAddDice(n_dice, n_trials=10000):
    sai=[]
    for i in range(n_trials):
        sai.append(sum(random.randint(1,6) for i in range(n_dice)))
        
    return sai

plt.hist(rollAndAddDice(2,10000), bins=np.arange(1, 13)+0.5, ec='black', rwidth=0.7)
plt.title('Histogram for Sum of Rolling 2 Dice')
plt.xlabel("Outcomes")
plt.ylabel("Frequency")
plt.show()
    
```
